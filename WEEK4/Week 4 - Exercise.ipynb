```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
framingham = pd.read_csv('framingham.csv')
```

```python
framingham.head()
```

```python
framingham.info()
```

```python
framingham.dropna(axis=0, inplace=True)
```

```python
X,y = framingham.drop('TenYearCHD', axis=1), framingham['TenYearCHD'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=92)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=4300)
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
model.score(X_test, y_test)
```
