Let's go back to our titanic example from the prevoius class.  We want to go through the same steps of reading in information, and preparing the data.  In this case I am going to use the built in data from seaborn.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

Drop the data we don't want.

```python
titanic.drop('embark_town', axis=1, inplace=True)
titanic.drop('deck', axis=1, inplace=True)
titanic.drop('embarked', axis=1, inplace=True)
```

```python
titanic.head()
```

Change data so that it is easeir to interpret.

```python
titanic.replace('male', 0, inplace=True)
titanic.replace('female', 1, inplace=True)
```

```python
titanic.head()
```

Do we have any trends?

```python
sns.pairplot(titanic)
```

### What factors might have effected survival?

```python
#Does sex play a role?
sns.barplot(x='sex', y='survived', data=titanic)
```

```python
#If you are by yourself?
sns.barplot(x='alone', y='survived', data=titanic)
```

```python
#If you are a male adult?
sns.barplot(x='adult_male', y='survived', data=titanic)
```

```python
#Do siblings or spouses matter?
sns.barplot(x='sibsp', y='survived', data=titanic)
```

```python
#Does your status on the ship matter?
sns.barplot(x='class', hue='who', y='survived', data=titanic)
```

```python
#What if we look at sibsp and class?
sns.barplot(x='sibsp', hue='class', y='survived', data=titanic)
```
